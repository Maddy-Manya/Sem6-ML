```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/advertising.csv", index_col="ID")
data
```

```python
def sales_regression_gradient_descent(feature):
    X = data[feature]
    y = data["Sales"]


    # starting params
    m = 0
    c = 0
    L = .0001 # ;earning param
    n = 1000 # iterations


    for i in range(n):
        y_pred = X*m + c

        D_m = -2/n * (X * ( y - y_pred)).sum()
        D_c = -2/n *  ( y - y_pred).sum()

        m -= L * D_m
        c -= L * D_c


    # plotting
    plt.scatter(X,y)
    plt.axline((0,c) , slope=m)
    plt.xlabel(feature)
    plt.ylabel("Sales")
    plt.title(f"Sales as function of {feature}")
```

```python
sales_regression_gradient_descent("TV")
```

```python
sales_regression_gradient_descent("Radio")
```

```python
sales_regression_gradient_descent("Newspaper")
```
